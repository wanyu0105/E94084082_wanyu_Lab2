```python
import pygame
import time

WIN_WIDTH = 1024
WIN_HEIGHT = 600
ENEMY_WIDTH = 50
ENEMY_HEIGHT = 55
BTN_WIDTH = 80
BTN_HEIGHT = 80
HP_WIDTH = 40
HP_HEIGHT = 40
FPS = 30

# color (RGB)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# initialization
pygame.init()
win = pygame.display.set_mode((WIN_WIDTH, WIN_HEIGHT))


# load image (background, enemy, buttons)
background_image = pygame.transform.scale(pygame.image.load("./images/Map.png"), (WIN_WIDTH, WIN_HEIGHT))
enemy_image = pygame.transform.scale(pygame.image.load("./images/enemy.png"), (ENEMY_WIDTH, ENEMY_HEIGHT))

hp_image = pygame.transform.scale(pygame.image.load("./images/hp.png"), (HP_WIDTH, HP_HEIGHT))
hp_gray_image = pygame.transform.scale(pygame.image.load("./images/hp_gray.png"), (HP_WIDTH, HP_HEIGHT))

muse_image = pygame.transform.scale(pygame.image.load("./images/muse.png"), (BTN_WIDTH, BTN_HEIGHT))
pause_image = pygame.transform.scale(pygame.image.load("./images/pause.png"), (BTN_WIDTH, BTN_HEIGHT))
sound_image = pygame.transform.scale(pygame.image.load("./images/sound.png"), (BTN_WIDTH, BTN_HEIGHT))
continue_image = pygame.transform.scale(pygame.image.load("./images/continue.png"), (BTN_WIDTH, BTN_HEIGHT))


# set the title
pygame.display.set_caption("My first game")

# clock
clock = pygame.time.Clock()

class Game:
    def __init__(self):
        # window
        self.win = win

        # hp
        self.hp = 7
        self.max_hp = 10
        pass

    def game_run(self):
        # game loop
        run = True
        while run:
            clock.tick(FPS)
            # event loop
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    run = False
                    
            # draw background
            win.blit(background_image, (0, 0))

            # draw enemy and health bar
            win.blit(enemy_image ,(55,255))
            pygame.draw.rect(win, RED , [55, 250 , 45, 5])
            
            # draw menu (and buttons)
            pygame.draw.rect(win, BLACK , [0, 0 , WIN_WIDTH, 80])
            win.blit(muse_image, (700, 0))
            win.blit(sound_image, (780, 0))
            win.blit(continue_image, (860, 0))
            win.blit(pause_image, (940, 0))
            
            # draw hp
            hp_position = 412   #第一張 hp 的圖所在的 x 座標
            hp_change_position = 0
            hp_change_position_2 = 0
             
            for i in range(10):   #利用for迴圈讓 hp 的圖一張接一張放上去
                if(i<5):
                    win.blit(hp_image, (hp_position + hp_change_position, 0 ))
                    hp_change_position += 40   #一張 hp 圖的長為 40
                elif(i<7):
                    win.blit(hp_image, (hp_position + hp_change_position_2 , 40))
                    hp_change_position_2 += 40
                else:
                    win.blit(hp_gray_image, (hp_position + hp_change_position_2 , 40))
                    hp_change_position_2 += 40


            # draw time
            game_font = pygame.font.SysFont(None, 24)   # font is used to render the text of the time onto a surface,and font set to size 24
            time_passed = pygame.time.get_ticks()   # 取得時間
            time_passed_seconds = time_passed//1000   # 將時間從毫秒換為秒
            # 利用 game_font.render() 設置 game time 顯示的時間與顏色
            game_time = game_font.render(str(time_passed_seconds//60)+":"+str(time_passed_seconds%60).zfill(2),True,WHITE)
            pygame.draw.rect(win, BLACK , [0, 573 , 50, 27])
            win.blit(game_time,(7,580))   
            
            pygame.display.update()
            
        pygame.quit()

if __name__ == "__main__":
    covid_game = Game()
    covid_game.game_run()


```
